```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
df.head(50)
```

```python
df['country'].fillna('unknown', inplace=True)
df['director'].fillna('unknown',inplace=True)
```

```python
sns.countplot(data=df, x='type', hue='type', palette='Set2', legend=False)
plt.title("Count of movies vs tv shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar', color='coral')
plt.title("top 10 countries by content")
plt.ylabel("Number of titles")
plt.xticks(rotation=45)
plt.show()
```

```python
df['release_year'] = df['release_year'].fillna(0).astype(int)
df['release_year'].value_counts().sort_index().plot(kind='line', marker='o', color='green')
plt.title("Content Released Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
from collections import Counter

genre_list = []
for i in df['listed_in']:
    genre_list.extend(i.split(', '))

top_genres = Counter(genre_list).most_common(10)
genres, counts = zip(*top_genres)

sns.barplot(x=list(counts), y=list(genres), hue=list(genres), palette='magma', legend=False)

plt.title("Top 10 Netflix Genres")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```
